```python
from groq import Groq
import json

client = client_groq
MODEL = 'llama3-groq-70b-8192-tool-use-preview'

def is_coastal(list_of_cities):
    try:
        cities = json.loads(list_of_cities)
        # assume coastal cities are hardcoded for simplicity
        coastal_cities = ['Budva', 'Sofia']
        result = [city for city in cities if city in coastal_cities]
        return json.dumps({"result": result})
    except:
        return json.dumps({"error": "Invalid expression"})

def weather(input_list):
    messages=[
        {
            "role": "system",
            "content": "You are a weather assistant. Use the is_coastal function to filter input list and provide the list of coastal cities as results."
        },
        {
            "role": "user",
            "content": json.dumps(input_list),
        }
    ]
    tools = [
        {
            "type": "function",
            "function": {
                "name": "is_coastal",
                "description": "filter list of cities to get only coastal cities",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "list_of_cities": {
                            "type": "array",
                            "description": "list of cities",
                        }
                    },
                    "required": ["list_of_cities"],
                },
            },
        }
    ]
    response = client.chat.completions.create(
        model=MODEL,
        messages=messages,
        tools=tools,
        tool_choice="auto",
        max_tokens=4096
    )

    response_message = response.choices[0].message
    tool_calls = response_message.tool_calls
    if tool_calls:
        available_functions = {
            "is_coastal": is_coastal,
        }
        messages.append(response_message)
        for tool_call in tool_calls:
            function_name = tool_call.function.name
            function_to_call = available_functions[function_name]
            function_args = json.loads(tool_call.function.arguments)
            function_response = function_to_call(
                list_of_cities=function_args.get("list_of_cities")
            )
            messages.append(
                {
                    "tool_call_id": tool_call.id,
                    "role": "tool",
                    "name": function_name,
                    "content": function_response,
                }
            )
        second_response = client.chat.completions.create(
            model=MODEL,
            messages=messages
        )
        return second_response.choices[0].message.content

input_list = json.dumps(['Budva', 'Podgorica', 'Belgrade', 'Sofia'])
print(weather(cities_list))
```
